```python
import pandas as pd
import os
```

```python
logs_df = pd.read_csv("../../data/memory_sample_kernel_log_round1_a_train.csv")
failure_df = pd.read_csv("../../data/memory_sample_failure_tag_round1_a_train.csv")
```

```python
# 转换日期时间格式
logs_df['collect_time'] = pd.to_datetime(logs_df['collect_time'], format='%Y-%m-%d %H:%M:%S')
failure_df['failure_time'] = pd.to_datetime(failure_df['failure_time'], format='%Y/%m/%d %H:%M')
```

```python
# 合并数据集
merged_data = pd.merge(logs_df, failure_df, on='serial_number', how='left')
```

```python
# 检查failure_time是否在collect_time的未来7天内
merged_data['target'] = (merged_data['failure_time'] - merged_data['collect_time']).dt.total_seconds() / (60 * 60 * 24)
merged_data['target'] = (merged_data['target'] >= 0) & (merged_data['target'] <= 7)
```

```python
# 将目标列转换为整数类型
merged_data['target'] = merged_data['target'].astype(int)

# 删除不再需要的列
merged_data = merged_data.drop(columns=['failure_time'])
```

```python
# 填充缺失值为0
merged_data.fillna(0, inplace=True)

# 再次显示数据集的前几行，确认修改
merged_data.head()
```
